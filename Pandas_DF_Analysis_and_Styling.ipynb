### Get Todays USD to EURO Exchange Rate

```python
import requests

url = 'https://api.exchangerate.host/convert?from=USD&to=EUR'
response = requests.get(url)
data = response.json()
```

### Style new DataFrame with different number of decimals as well as display prices in USD and EURO based on Exchange Rate.

```python
exchrate = data['info']['rate']
print(exchrate)
cols=['Open','High','Low','Close']
addcols=['OpenEuro','HighEuro','LowEuro','CloseEuro']
df_new = df[cols].head(10)

df_new['OpenEuro'] = df['Open'] * exchrate
df_new['HighEuro'] = df['High'] * exchrate
df_new['LowEuro'] = df['Low'] * exchrate
df_new['CloseEuro'] = df['Close'] * exchrate

eurosymbol = u"\N{euro sign}"

(df_new
 .style
 .format({"Open": "${:20,.2f}", 
          "High": "${:20,.3f}", 
          "Low": "${:20,.4f}",
          "Close":"${:20,.5f}",
          "OpenEuro": eurosymbol + "{:20,.5f}", 
          "HighEuro": eurosymbol + "{:20,.4f}",  
          "LowEuro": eurosymbol + "{:20,.3f}",
          "CloseEuro":eurosymbol + "{:20,.2f}"})
)
```

```python
df_new.describe()
```

### The format function has a `precision` argument to specifically help formatting floats, as well as decimal and thousands separators to support other locales, an `na_rep` argument to display missing data

```python
def style_missing_data(v, props=''):
    return props if np.isnan(v) else None

df_large = df_new + 1000000
df_large.iloc[3:6,2:4] = np.nan
(
df_large.head(6)
   .style
   .applymap(style_missing_data, props='background-color: black; color:yellow; font-size:150%;')
   .format(precision=3, thousands=",", na_rep = 'Missing Data')
)
```

### Set css style values for headers, rows, cells

Note: overwrite: (bool), default True
Styles are replaced if True, or extended if False. CSS rules are preserved so most recent styles set will dominate if selectors intersect.

```python
(
df_new.head(10)
  .style
     .set_table_styles([  # create internal CSS classes
        {'selector': 'th.col_heading', 'props': 'background-color: yellow;'},
        {'selector': 'th.col_heading.level0', 'props': 'background-color: white;'},
        {'selector': '.index_name', 'props': 'background-color: darkgrey;'},
        {'selector': 'tr:nth-child(5)', 'props': 'background-color: grey;color:blue;'},
        {'selector': 'tr:nth-child(5) td:nth-child(5)', 'props': 'background-color: black; color:yellow; font-size:150%;'},
        {'selector': 'tr', 'props': 'background-color: lightgrey;color:blue;'}],
        overwrite=False)
)
```

### Get min and max values from selected columns in dataframe

```python
cols=['Open','High','Low','Close']
df_min = df_new[cols].min()
df_max = df_new[cols].max()
df_median = df_new.median()

vmin = df_min.min()
vmax = df_max.max()
vmedian = df_median.median()

print(type(vmedian))
```

### Stle dataframe according to min and max price values

```python

cols=['Open','High','Low','Close']
df_new[cols].head(10).style.background_gradient(cmap='YlOrRd', vmin=vmin, vmax=vmax)
```

### Show stock prices below $114 in bold red

```python
print(vmedian)

def style_lowprice(v, props=''):
    return props if v < vmedian else None

( df.iloc[10:20,1:3].style.applymap(style_lowprice, props='color:red;font-weight:bold; font-size:120%')
              .applymap(lambda v: 'opacity: 80%;' if (v < vmedian) else None)
)
```

### Highlight max stock value by row comparing Open, High, Low, Close and format stock prices with two decimals

```python
def highlight_max(s, props=''):
    return np.where(s == np.nanmax(s.values), props, '')

(df.iloc[80:90,0:4]
     .style
     .format('{:.2f}')
     .apply(highlight_max, props='color:white;background-color:green', axis=1)
)
```

### Do some css styling for headers, index_names,  rows, and also create hover effect for each row

```python
cell_hover = {  # for row hover use <tr> instead of <td>
    'selector': 'td:hover',
    'props': [('background-color', 'white'),('color', 'blue')]
}
index_names = {
    'selector': '.index_name',
    'props': 'font-size: 120%; text-alignment: center; background-color: #6b0000; color: orange; font-weight:normal;'
}
headers = {
    'selector': 'th:not(.index_name)',
    'props': 'background-color: #883232; color: lightgreen; text-align:center;'
}

(df.iloc[50:60,0:4]
    .style
    .format(precision=2)
    .set_table_styles([cell_hover, index_names, headers])
)
```
